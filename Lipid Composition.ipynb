```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.colors import ListedColormap
```

```python
pal_hls = sns.hls_palette(20, l=0.7, s=1, h=1).as_hex(); print(pal_hls)
colors = {'CHL': pal_hls[0],'PSM 16:0-18:1': pal_hls[6],'PLPC 16:0-18:2': pal_hls[11],'SAPS 18:0-20:4': pal_hls[1],'PLA20 18:1-20:4': pal_hls[5],'PAPS 16:0-20:4': pal_hls[13],'PDOPE 16:0-22:6': pal_hls[2],'POPC 16:0-18:1': pal_hls[12],'SAPE 18:0-20:4': pal_hls[7],'POPE 16:0-18:1': pal_hls[3],'DPPC 16:0-16:0': pal_hls[9],'SAPI2 18:1-20:4': pal_hls[14],'PAPC 16:0-20:4': pal_hls[4], 'NSM 24:1-18:1': pal_hls[8], 'SOPC 18:0-18:1': pal_hls[15],'LSM 24:0-18:1': pal_hls[10]}
```

```python
lipids_i=["PAPS 16:0-20:4", "PLPC 16:0-18:2", "POPC 16:0-18:1", "PLA20 18:1-20:4", "POPE 16:0-18:1", "PDOPE 16:0-22:6", "SAPE 18:0-20:4", "DPPC 16:0-16:0", "PSM 16:0-18:1", "SAPS 18:0-20:4", "SAPI2 18:1-20:4", "CHL"]
lipids_o=["PSM 16:0-18:1", "PLPC 16:0-18:2", "NSM 24:1-18:1", "LSM 24:0-18:1", "SOPC 18:0-18:1", "PAPC 16:0-20:4","SAPS 18:0-20:4", "PLA20 18:1-20:4", "CHL"]
len(lipids_o)
```

```python
colors_i = [colors[lipid] for lipid in lipids_i]
colors_o = [colors[lipid] for lipid in lipids_o]
```

### Inner Leaflet

```python
sns.set_style("whitegrid", {'font.family':'serif', 'font.serif':'Computer Modern'})
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
lipids_i=["PAPS 16:0-20:4", "PLPC 16:0-18:2", "POPC 16:0-18:1", "PLA20 18:1-20:4", "POPE 16:0-18:1", "PDOPE 16:0-22:6", "SAPE 18:0-20:4", "DPPC 16:0-16:0", "PSM 16:0-18:1", "SAPS 18:0-20:4", "SAPI2 18:1-20:4", "CHL"]
comp_inner = [19,13,5,17,3,11,5,3,2,2,3,59 ]
ax.pie(comp_inner, labels=lipids_i , autopct='%1.0f%%', colors=colors_i) 
plt.title("Lipid Compositions in the inner leaflet")

plt.savefig("lipid_comp_inner.svg", bbox_inches='tight')
plt.plot()
```
